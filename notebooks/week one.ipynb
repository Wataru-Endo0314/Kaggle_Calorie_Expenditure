```python
# データ読み込み
path = '../input'
train = pd.read_csv(path+'/train.csv')
test = pd.read_csv(path+'/test.csv')
```

## EDA

```python
sns.boxplot(data = train.drop(columns = 'id'))
```

```python
sns.boxplot(data = test.drop(columns = 'id'))
```

```python
sns.boxplot(data = train['Body_Temp'])
```

```python
sns.boxplot(data = test['Body_Temp'])
```

```python
# 性別とカロリー消費の関係

sns.boxplot(x=train['Sex'], y = train['Calories'])
```

```python
# 年齢とカロリーの関係
plt.figure(figsize=(15, 5))
sns.boxplot(x=train['Age'], y = train['Calories'])
plt.tight_layout()
```

```python
# そもそも年代としてはどこの年代が多いのか
plt.figure(figsize=(15, 5))
sns.countplot(x=train['Age'])
plt.tight_layout()
```

```python
plt.figure(figsize=(15, 5))
sns.countplot(x=test['Age'])
plt.tight_layout()
```

```python
# 心拍数と消費カロリー
sns.lmplot(x='Heart_Rate', y='Calories', hue='Sex', data=train)
plt.tight_layout()
```
